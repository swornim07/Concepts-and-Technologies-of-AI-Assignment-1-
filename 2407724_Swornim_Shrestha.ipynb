## Task- 3- Outlier Detection:
 ### Tasks:
 1. Identify outlier countries in South Asia based on their Score and GDP per Capita.
 2. Define outliers using the 1.5 × IQR rule.
 3. Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers
 in a different color.
 4. Discuss the characteristics of these outliers and their potential impact on regional averages.

```python
# 1. identifying Outliers using the 1.5 × IQR rule
# for Score and Log GDP per capita

# calculating the IQR (Interquartile Range) for Score and Log GDP per capita
score_q1 = south_asia_data['score'].quantile(0.25)
score_q3 = south_asia_data['score'].quantile(0.75)
score_iqr = score_q3 - score_q1

gdp_q1 = south_asia_data['Log GDP per capita'].quantile(0.25)
gdp_q3 = south_asia_data['Log GDP per capita'].quantile(0.75)
gdp_iqr = gdp_q3 - gdp_q1

# defining the outlier thresholds
score_lower_bound = score_q1 - 1.5 * score_iqr
score_upper_bound = score_q3 + 1.5 * score_iqr

gdp_lower_bound = gdp_q1 - 1.5 * gdp_iqr
gdp_upper_bound = gdp_q3 + 1.5 * gdp_iqr

# filtering out the outliers based on these thresholds
outliers_score = south_asia_data[(south_asia_data['score'] < score_lower_bound) |
                                 (south_asia_data['score'] > score_upper_bound)]

outliers_gdp = south_asia_data[(south_asia_data['Log GDP per capita'] < gdp_lower_bound) |
                               (south_asia_data['Log GDP per capita'] > gdp_upper_bound)]

# combining the outliers from both columns
outliers = pd.concat([outliers_score, outliers_gdp]).drop_duplicates()

# 2. creating a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Log GDP per capita', y='score', data=south_asia_data, color='red', label='Normal Data')
sns.scatterplot(x='Log GDP per capita', y='score', data=outliers, color='blue', label='Outliers')

plt.title('Scatter Plot: GDP per Capita vs Score with Outliers Highlighted')
plt.xlabel('Log GDP per Capita')
plt.ylabel('Happiness Score')
plt.legend()
plt.show()

# 3. displaying the outliers for inspection
print("Outlier Countries Based on Score and GDP per Capita:")
print(outliers[['Country name', 'score', 'Log GDP per capita']])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# checking column names to make sure we are using the correct ones
print("South Asia Data Columns:")
print(south_asia_data.columns)

print("\nMiddle East Data Columns:")
print(middle_east_data.columns)

# creating Function to detect outliers using the 1.5 * IQR rule
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# 1. detecting outliers in South Asia based on Score and GDP per Capita
south_asia_outliers_score = detect_outliers(south_asia_data, 'score')  # replace 'score' with correct column name if necessary
south_asia_outliers_gdp = detect_outliers(south_asia_data, 'Log GDP per capita')  # replace with correct column name if necessary

# 2. detecting outliers in Middle East based on Score and GDP per Capita
middle_east_outliers_score = detect_outliers(middle_east_data, 'score')  # replace 'score' with correct column name if necessary
middle_east_outliers_gdp = detect_outliers(middle_east_data, 'Log GDP per capita')  # replace with correct column name if necessary

# 3. combining outliers for visualization
south_asia_outliers = pd.concat([south_asia_outliers_score, south_asia_outliers_gdp]).drop_duplicates()
middle_east_outliers = pd.concat([middle_east_outliers_score, middle_east_outliers_gdp]).drop_duplicates()

# 4. scatter plotting for South Asia: Score vs GDP per Capita
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Log GDP per capita', y='score', data=south_asia_data, label="Normal Data")  # replace with correct column name if necessary
sns.scatterplot(x=south_asia_outliers['Log GDP per capita'], y=south_asia_outliers['score'], color='green', label="Outliers")
plt.title("South Asia: Score vs GDP per Capita (Outliers in Green)")
plt.legend()
plt.show()

# 5. scatter plotting for Middle East: Score vs GDP per Capita
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Log GDP per capita', y='score', data=middle_east_data, label="Normal Data")  # replace with correct column name if necessary
sns.scatterplot(x=middle_east_outliers['Log GDP per capita'], y=middle_east_outliers['score'], color='green', label="Outliers")
plt.title("Middle East: Score vs GDP per Capita (Outliers in Green)")
plt.legend()
plt.show()

# displaying outliers for both regions
print("South Asia Outliers (Score and GDP per Capita):")
print(south_asia_outliers)

print("\nMiddle East Outliers (Score and GDP per Capita):")
print(middle_east_outliers)
```
